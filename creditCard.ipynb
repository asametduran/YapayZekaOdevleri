```python
!pip install minisom
import pandas as pd
from minisom import MiniSom
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
veri = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/YapayZekaOdevleri/CC GENERAL.csv')
veri.dropna()
X = veri[['BALANCE', 'PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT']]
```

```python
kmeans = KMeans(n_clusters=4, max_iter=300, random_state=42)
# Fit the KMeans model to the data
kmeans.fit(X[['BALANCE', 'PURCHASES']]) # Fit using the relevant columns
plt.scatter(X['BALANCE'], X['PURCHASES'], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.show()
```

```python

veri["kümeler"] = kmeans.labels_
print(veri[veri["kümeler"] == 0].head(5))
print(veri[veri["kümeler"] == 1].head(5))
print(veri[veri["kümeler"] == 2].head(5))
```
