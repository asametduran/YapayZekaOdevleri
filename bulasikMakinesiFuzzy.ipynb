```python
!pip install scikit-fuzzy
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
bulaşık_miktarı = ctrl.Antecedent(np.arange(0, 11, 1), 'bulaşık_miktarı')  # 0-10 arası (az - çok)
kirlilik_derecesi = ctrl.Antecedent(np.arange(0, 11, 1), 'kirlilik_derecesi')  # 0-10 arası (temiz - kirli)
bulaşık_cinsi = ctrl.Antecedent(np.arange(0, 3, 1), 'bulaşık_cinsi')  # 0: hassas, 1: karma, 2: güçlü
```

```python
yıkama_zamanı = ctrl.Consequent(np.arange(0, 181, 1), 'yıkama_zamanı')  # 0-180 dakika
deterjan_miktarı = ctrl.Consequent(np.arange(0, 101, 1), 'deterjan_miktarı')  # 0-100% oranında
su_sıcaklığı = ctrl.Consequent(np.arange(30, 101, 1), 'su_sıcaklığı')  # 30-100°C
üst_pompa_devri = ctrl.Consequent(np.arange(0, 4001, 1), 'üst_pompa_devri')  # 0-4000 dev/dak
alt_pompa_devri = ctrl.Consequent(np.arange(0, 4001, 1), 'alt_pompa_devri')
```

```python
bulaşık_miktarı['az'] = fuzz.trimf(bulaşık_miktarı.universe, [0, 0, 5])
bulaşık_miktarı['orta'] = fuzz.trimf(bulaşık_miktarı.universe, [0, 5, 10])
bulaşık_miktarı['çok'] = fuzz.trimf(bulaşık_miktarı.universe, [5, 10, 10])
```

```python
kirlilik_derecesi['temiz'] = fuzz.trimf(kirlilik_derecesi.universe, [0, 0, 5])
kirlilik_derecesi['orta'] = fuzz.trimf(kirlilik_derecesi.universe, [0, 5, 10])
kirlilik_derecesi['kirli'] = fuzz.trimf(kirlilik_derecesi.universe, [5, 10, 10])
```

```python
bulaşık_cinsi['hassas'] = fuzz.trimf(bulaşık_cinsi.universe, [0, 0, 1])
bulaşık_cinsi['karma'] = fuzz.trimf(bulaşık_cinsi.universe, [0, 1, 2])
bulaşık_cinsi['güçlü'] = fuzz.trimf(bulaşık_cinsi.universe, [1, 2, 2])
```

```python
yıkama_zamanı['kısa'] = fuzz.trimf(yıkama_zamanı.universe, [0, 0, 60])
yıkama_zamanı['orta'] = fuzz.trimf(yıkama_zamanı.universe, [30, 60, 120])
yıkama_zamanı['uzun'] = fuzz.trimf(yıkama_zamanı.universe, [60, 120, 180])
```

```python
deterjan_miktarı['az'] = fuzz.trimf(deterjan_miktarı.universe, [0, 0, 50])
deterjan_miktarı['orta'] = fuzz.trimf(deterjan_miktarı.universe, [25, 50, 75])
deterjan_miktarı['çok'] = fuzz.trimf(deterjan_miktarı.universe, [50, 100, 100])
```

```python
su_sıcaklığı['düşük'] = fuzz.trimf(su_sıcaklığı.universe, [30, 30, 50])
su_sıcaklığı['orta'] = fuzz.trimf(su_sıcaklığı.universe, [40, 60, 80])
su_sıcaklığı['yüksek'] = fuzz.trimf(su_sıcaklığı.universe, [70, 90, 100])
```

```python
üst_pompa_devri['düşük'] = fuzz.trimf(üst_pompa_devri.universe, [0, 0, 2000])
üst_pompa_devri['orta'] = fuzz.trimf(üst_pompa_devri.universe, [1000, 2000, 3000])
üst_pompa_devri['yüksek'] = fuzz.trimf(üst_pompa_devri.universe, [2000, 3000, 4000])
```

```python
alt_pompa_devri['düşük'] = fuzz.trimf(alt_pompa_devri.universe, [0, 0, 2000])
alt_pompa_devri['orta'] = fuzz.trimf(alt_pompa_devri.universe, [1000, 2000, 3000])
alt_pompa_devri['yüksek'] = fuzz.trimf(alt_pompa_devri.universe, [2000, 3000, 4000])
```

```python
kural1 = ctrl.Rule(bulaşık_miktarı['az'] & kirlilik_derecesi['temiz'] & bulaşık_cinsi['hassas'],
                   yıkama_zamanı['kısa'])

kural2 = ctrl.Rule(bulaşık_miktarı['orta'] & kirlilik_derecesi['orta'] & bulaşık_cinsi['karma'],
                   yıkama_zamanı['orta'])

kural3 = ctrl.Rule(bulaşık_miktarı['çok'] & kirlilik_derecesi['kirli'] & bulaşık_cinsi['güçlü'],
                   yıkama_zamanı['uzun'])
```

```python
kontrol_sistemi = ctrl.ControlSystem([kural1, kural2, kural3])
simülasyon = ctrl.ControlSystemSimulation(kontrol_sistemi)
```

```python
simülasyon.input['bulaşık_miktarı'] = 8  # Örnek giriş
simülasyon.input['kirlilik_derecesi'] = 7
simülasyon.input['bulaşık_cinsi'] = 2  # Güçlü bulaşık

simülasyon.compute()
```
