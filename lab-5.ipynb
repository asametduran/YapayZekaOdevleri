```python
!pip install minisom
import pandas as pd
from minisom import MiniSom      # This should work now that SimpSOM is installed
from sklearn.cluster import KMeans
import numpy as np
```

Veri setinin yüklenmesi ve gerekli veriler ile kullanılması

```python
veri = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/YapayZekaOdevleri/airline-safety.csv")
X = veri.drop(["airline","avail_seat_km_per_week"],axis = 1)
```

Ağın oluşturulması

```python
som = MiniSom(20, 20, X.shape[1], sigma=1.0, learning_rate=0.5)
som.train_random(X.values, 10000)
```

Veri noktalarının 2 boyutlu bir haritaya gömülmesi ve kümlemenin yapılması

```python
hrt = np.array([som.winner(x) for x in X.values])
kmeans = KMeans(n_clusters=3, max_iter=300,random_state=0)
```

Kümeleme sonuçlarının gösterilmesi

```python
y_kmeans = kmeans.fit_predict(hrt)
print(y_kmeans)
```

Kümelerin etiketlerinin belirlenmesi

```python
veri["kümeler"] = kmeans.labels_
```

```python
print(veri[veri["kümeler"]==0].head(5))
```
