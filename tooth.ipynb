```python
# Uzman sistem için kuralları ve çözümleri tanımlayalım
def diş_hastalıkları():
    print("Diş ağrısı için uzman sistemine hoş geldiniz!")

    # Kullanıcıdan verileri alalım
    diş_fırçalarken_kanama = input("Diş fırçalarken kanama oldu mu? (evet/hayır): ").lower()
    uzun_süreli_kanama = input("Diş fırçalarken uzun süreli kanama oldu mu? (evet/hayır): ").lower()
    diş_eti_çekilmesi = input("Diş eti çekilmesi var mı? (evet/hayır): ").lower()
    diş_kökü_görünümü = input("Diş kökü görünüyorsa? (evet/hayır): ").lower()
    renk_değişimi = input("Dişte yiyecek ve içeceklerden dolayı renk değişimi var mı? (evet/hayır): ").lower()
    yeni_diș_çıkışı = input("Yeni diş çıkarken morarma var mı? (evet/hayır): ").lower()
    çürük_var = input("Dişte ağrı yapmayan çürük var mı? (evet/hayır): ").lower()
    çürük_ileri_düzey = input("Dişteki çürük ileri derecede mi? (evet/hayır): ").lower()

    # Kurallara göre kararları verelim
    if diş_fırçalarken_kanama == 'evet':
        print("Diş hastalığı vardır, diş hekimine başvur.")
    elif uzun_süreli_kanama == 'evet':
        print("Dişeti çekilmesi vardır, diş hekimine başvur.")
    elif diş_eti_çekilmesi == 'evet' and diş_kökü_görünümü == 'evet':
        print("Diş kökü görünüyorsa, dolgu yaptır.")
    elif renk_değişimi == 'evet':
        print("Dişlerinizi temizleyin.")
    elif yeni_diș_çıkışı == 'evet':
        print("Yeni diş çıkarken morarma görünüyor, diş hekimine başvur.")
    elif çürük_var == 'evet':
        print("Dişte ağrı yapmayan çürük var, dolgu yaptır.")
    elif çürük_ileri_düzey == 'evet':
        print("Dişteki çürük ileri derecedeyse, kanal tedavisi ve dolgu yaptır.")
    else:
        print("Diş sağlığınızda belirgin bir sorun görünmüyor, ancak düzenli kontrolleri ihmal etmeyin.")

# Fonksiyonu çalıştırarak uzman sistemi başlatalım
diş_hastalıkları()
```
