```python
!pip install minisom
import pandas as pd
from minisom import MiniSom
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
veri = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/YapayZekaOdevleri/Mall_Customers.csv")
```

```python
X = veri[['Age','Annual Income (k$)']]
```

```python
kmeans = KMeans(n_clusters = 3, max_iter=300,random_state=42)
```

```python
veri['kümeler'] = kmeans.fit_predict(X)
```

```python
plt.scatter(X['Age'],X['Annual Income (k$)'],c=veri['kümeler'],cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.title("Customer Segmentation")
plt.colorbar(label="Cluster Number")
plt.show()
```
